```python
from flask import Flask, jsonify
import pandas as pd
import time
import threading

app = Flask(__name__)
data_values = []
current_index = [0]

# Load CSV values at startup
df = pd.read_csv("PCA1.csv")
data_values = df.iloc[:, 0].tolist()  # assuming values are in the first column

# Polling function that simulates real-time value generation
def stream_data():
    while True:
        if current_index[0] < len(data_values) - 1:
            time.sleep(2.5)
            current_index[0] += 1
        else:
            break

@app.route('/get-value', methods=['GET'])
def get_value():
    value = data_values[current_index[0]]
    return jsonify({'value': value})

if __name__ == '__main__':
    threading.Thread(target=stream_data, daemon=True).start()
    app.run(host='0.0.0.0', port=5000)

```
